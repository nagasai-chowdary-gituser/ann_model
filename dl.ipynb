```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv(r'C:\Users\asus\Desktop\aidsml_engineering\AiMlDs_workspace\deep_learning\Churn_Modelling.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
```

```python
ohe=OneHotEncoder()
geo_hot=ohe.fit_transform(df[['Geography']]).toarray()  
```

```python
geo_hot_df=pd.DataFrame(geo_hot,columns=ohe.get_feature_names_out())
```

```python
df=pd.concat([df,geo_hot_df],axis=1)
```

```python
df.drop(['Geography'],axis=1,inplace=True)
```

```python
## Save the encoders and sscaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(le,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x
```

```python
x.drop(['Geography_France','Geography_Germany','Geography_Spain'],axis=1,inplace=True)
```

```python
x=pd.concat([x,geo_hot_df],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=22)
std=StandardScaler()
x_train=std.fit_transform(x_train)
x_test=std.transform(x_test)
```

```python
with open("scaler.pkl",'wb') as file:
    pickle.dump(std,file)
```

```python
import tensorflow
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import datetime
```

```python
(x_train.shape[1],)
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
model.compile(optimizer=opt,loss='BinaryCrossentropy',metrics=['accuracy'])
```

```python
logss="logs/fit/"+datetime.datetime.now().strftime("%y%m%d-%H%M%S")
tensor_callback=TensorBoard(log_dir=logss,histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensor_callback,early_stopping_callback]
)
```

```python
model.save("model.h5")
```
