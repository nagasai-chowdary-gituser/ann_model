```python
import tensorflow
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
model=load_model('model.h5')
with open('onehot_encoder_geo.pkl','rb') as file:
    one_enc_geo=pickle.load(file)
with open('label_encoder_gender.pkl','rb') as file:
    label_enc=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
geo_hot=one_enc_geo.transform([[input_data['Geography']]]).toarray()
geo_hot_df=pd.DataFrame(geo_hot,columns=one_enc_geo.get_feature_names_out(['Geography']))
geo_hot_df
```

```python
input_data
```

```python
input_data_df=pd.DataFrame([input_data])
input_data_df
```

```python
input_data_df['Gender']=label_enc.transform([input_data_df['Gender']])
input_data_df
```

```python
input_data_df=pd.concat([input_data_df.drop('Geography',axis=1),geo_hot_df],axis=1)
```

```python
 input_scaled=scaler.transform(input_data_df)
 input_scaled
```

```python
predictions=model.predict(input_scaled)
```

```python
prediction_prob=predictions[0][0]
prediction_prob
```

```python
if prediction_prob>0.5:
    print("customer is likely to churn")
else:
    print("not likely to churn")
```
